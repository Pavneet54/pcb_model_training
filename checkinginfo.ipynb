```python

from pathlib import Path
import cv2
import numpy as np

mask_dirs = [
    Path("PCBSegClassNet/data/segmentation/train/masks"),
    Path("PCBSegClassNet/data/segmentation/val/masks")
]

classes = set()

for mask_dir in mask_dirs:
    for mask_file in mask_dir.glob("*"):
        if mask_file.suffix.lower() in [".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff"]:
            mask = cv2.imread(str(mask_file), cv2.IMREAD_GRAYSCALE)

            if mask is None:
                continue

            classes.update(np.unique(mask))

print("="*40)
print("Total Classes Found:", len(classes))
print("Class IDs:", sorted(classes))
```

```python
from pathlib import Path
import cv2
import numpy as np
from collections import defaultdict

mask_dirs = [
    Path("PCBSegClassNet/data/segmentation/train/masks"),
    Path("PCBSegClassNet/data/segmentation/val/masks")
]

class_image_count = defaultdict(int)
class_pixel_count = defaultdict(int)

for mask_dir in mask_dirs:
    for mask_file in mask_dir.glob("*.png"):   # change if masks are jpg/jpeg
        mask = cv2.imread(str(mask_file), cv2.IMREAD_GRAYSCALE)

        unique_classes = np.unique(mask)

        for cls in unique_classes:
            class_image_count[int(cls)] += 1
            class_pixel_count[int(cls)] += np.sum(mask == cls)

print(f"{'Class':<8}{'Images':<12}{'Pixels'}")
print("-" * 35)

for cls in sorted(class_image_count.keys()):
    print(f"{cls:<8}{class_image_count[cls]:<12}{class_pixel_count[cls]}")
```

```python
from pathlib import Path
import cv2
import numpy as np

mapping = {
0:0,
66:1,
67:2,
68:3,
76:4,
79:5,
84:6,
108:7,
111:8,
115:9,
116:10,
150:11,
155:12,
166:13,
181:14,
200:15,
201:16,
204:17,
206:18,
208:19,
219:20,
222:21,
225:22,
226:23,
245:24
}

for mask_dir in mask_dirs:

    output_dir = mask_dir.parent/"remapped_masks"
    output_dir.mkdir(exist_ok=True)

    for mask_file in mask_dir.glob("*.png"):

        mask = cv2.imread(str(mask_file),0)

        remapped = np.zeros_like(mask)

        for old,new in mapping.items():
            remapped[mask==old]=new

        cv2.imwrite(str(output_dir/mask_file.name),remapped)

print("Finished.")
```

```python
from pathlib import Path
import cv2
import numpy as np

# Path to remapped masks
mask_dir = Path("PCBSegClassNet/data/segmentation/train/remapped_masks")

# Number of classes after remapping
NUM_CLASSES = 25

# Read the first mask (for demonstration)
mask_path = next(mask_dir.glob("*.png"))

print("Processing:", mask_path.name)

mask = cv2.imread(str(mask_path), cv2.IMREAD_GRAYSCALE)

print("Mask Shape:", mask.shape)

# Extract contours for each class
for class_id in range(NUM_CLASSES):

    # Create binary mask for one class
    binary = np.uint8(mask == class_id) * 255

    # Find contours
    contours, _ = cv2.findContours(
        binary,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )

    print(f"Class {class_id}: {len(contours)} contour(s)")
```

```python
from pathlib import Path
import cv2
import numpy as np

mask_dir = Path("PCBSegClassNet/data/segmentation/train/remapped_masks")

mask_path = next(mask_dir.glob("*.png"))

mask = cv2.imread(str(mask_path), cv2.IMREAD_GRAYSCALE)

NUM_CLASSES = 25

for class_id in range(1, NUM_CLASSES):      # Skip background (0)

    binary = np.uint8(mask == class_id) * 255

    contours, _ = cv2.findContours(
        binary,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )

    print(f"\nClass {class_id}")

    for i, contour in enumerate(contours):

        epsilon = 0.002 * cv2.arcLength(contour, True)

        polygon = cv2.approxPolyDP(
            contour,
            epsilon,
            True
        )

        print(f"Object {i+1}")
        print("Original contour points :", len(contour))
        print("Polygon vertices        :", len(polygon))
```
